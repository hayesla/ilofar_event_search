```python
from sunpy.net import hek
import pandas as pd 
import time
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import urljoin
from sunpy.time import parse_time
import numpy as np
from pathlib import Path
import datetime
import urllib
client = hek.HEKClient()
```

```python
def find_lofar_files(date):
    """
    Find I-LOFAR dynamic spectra for date given.
    
    Parameters
    ----------
    date : ~`datetime.datetime` or ~`astropy.Time`
    
    Returns
    -------
    `list` of available files
    """
    baseurl = "https://data.lofar.ie/{:s}/bst/kbt/rcu357_1beam/".format(date.strftime("%Y/%m/%d"))
    file_urls = []
    try:
        opn = urlopen(baseurl)
        try:
            soup = BeautifulSoup(opn, "html.parser")
            for link in soup.find_all("a"):
                href = link.get("href")
                if href.endswith("00X.dat"):# or href.endswith("00Y.dat"):
                    file_urls.append(urljoin(baseurl, href))
        except:
            print("No file of interest")
    except:
        print("No page")
    return file_urls
```

```python
tstart = parse_time('2017-09-01').datetime
tend = parse_time('2020-10-18').datetime
```

```python
all_days = []
day1 = tstart
all_days.append(day1)
while day1 <= tend:
	day1 = day1 + datetime.timedelta(days=1)
	all_days.append(day1)
```

```python
files = []
for d in all_days:
    f = find_lofar_files(d)
    files.append(f)
```

```python
def get_urls(url_list, path='./bst_files/'):
    
    
    for files in url_list:
        if len(files)>1:
            for f in files:
        
                if Path(path).joinpath(f.split("/")[-1]).exists():
                    print("{:s} file already exists in given path".format(f.split("/")[-1]))
                    continue 
                urllib.request.urlretrieve(f, Path(path).joinpath(f.split("/")[-1]))

                if Path(path).joinpath(f.split("/")[-1]).exists():
                    print("download success")
        else:
            f = files[0]
            if Path(path).joinpath(f.split("/")[-1]).exists():
                print("{:s} file already exists in given path".format(f.split("/")[-1]))
                continue 
            urllib.request.urlretrieve(f, Path(path).joinpath(f.split("/")[-1]))

            if Path(path).joinpath(f.split("/")[-1]).exists():
                print("download success")            
```
